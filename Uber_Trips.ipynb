```python
import pandas as pd

url = "https://raw.githubusercontent.com/fivethirtyeight/uber-tlc-foil-response/master/uber-trip-data/uber-raw-data-apr14.csv"

pdf = pd.read_csv(url)
```

```python
pdf['Date/Time'] = pd.to_datetime(pdf['Date/Time'])
```

Extract Hour, Day and Month

```python
pdf['Hour'] = pdf['Date/Time'].dt.hour
pdf['Day'] = pdf['Date/Time'].dt.day
pdf['Month'] = pdf['Date/Time'].dt.month
```

 Trips by Hour

```python
hourly = pdf.groupby('Hour').size()

print(hourly)
```

```python
# @title Default title text
import matplotlib.pyplot as plt

hourly.plot(kind='bar')
plt.xlabel('Hour')
plt.ylabel('Number of Trips')
plt.title('Uber Trips by Hour')
plt.show()
```

 Plot Hourly Trips

```python
daily = pdf.groupby('Day').size()

daily.plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Number of Trips')
plt.title('Uber Trips by Day')
plt.show()
```

 Trips by Month

```python
monthly = pdf.groupby('Month').size()

monthly.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.title('Uber Trips by Month')
plt.show()
```

 Find Peak Hour

```python
peak_hour = hourly.idxmax()
max_rides = hourly.max()

print("Peak Hour:", peak_hour)
print("Number of Trips:", max_rides)
```
